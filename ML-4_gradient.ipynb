```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+3) ** 2
    
def grad_f(x):
    return 2 * (x + 3)  #derivative/slope of f(x)
```

```python
x_curr=2 #start pt
learning_rate = 0.1 #step size
tolerance = 1e-6 #convergence/stop pt
max_iters = 25  #max iters
history = [x_curr] #store history of x
```

```python
for i in range(max_iters):
    gradient = grad_f(x_curr)
    x_next= x_curr - learning_rate * gradient  #update/next step

    # Check convergence
    if abs(x_next - x_curr) < tolerance:
        print(f"Converged after {i+1} iterations.")
        break
        
    x_curr=x_next
    history.append(x_curr)
    print(f"Iteration {i+1}: x= {x_curr: .4f}, f(x) = {f(x_curr): .4f}")
```

```python
print("Local minima at x= ", x_curr)
print("Function val at local minima y = ", f(x_curr))
```

```python
plt.plot(history, [f(val) for val in history], marker='o')
plt.xlabel("x values")
plt.ylabel("f(x)")
plt.title("Gradient Descent Convergence")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# function and derivative
def f(x): return (x + 3)**2
def df(x): return 2 * (x + 3) #derivative(slope)

x = 2  #start pt
lr = 0.1 #step
epochs = 20

x_history = [x]
for i in range(epochs):
    x = x - lr * df(x)
    x_history.append(x)
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

# Step 4: Print result
print("\nLocal minima occurs at x =", round(x, 4))
print("Minimum value of function =", round(f(x), 4))

# simple plot
plt.plot([f(i) for i in x_history], marker='o')
plt.title("Gradient Descent Steps")
plt.xlabel("Step")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
